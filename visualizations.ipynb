```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("india_population.csv")
df["Population"] = (df["Population"])/1000000000
df
```

```python
a = df["Year"] 
b = df["Population"]
```

```python
plt.plot(a,b,color="violet",lw=4)
plt.xlabel("year")
plt.ylabel("population in billions")
plt.show()
```

```python
plt.plot(a,b,marker="*",mew=10,color="y")
```

```python
plt.plot(a,b,color="red",lw=4,marker="o",mew=5)
plt.xlabel("year")
plt.ylabel("population in billions")
plt.show()
```

```python
plt.scatter(a,b)
```

```python
plt.boxplot(b)
```

```python
plt.hist(b)
```
